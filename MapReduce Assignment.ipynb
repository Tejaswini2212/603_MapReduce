```python
import PyPDF2
import math

def extract_pages_from_pdf(pdf_path, start_page, num_pages, output_file):
    with open(pdf_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        text = ""

        # Ensuring start_page is valid
        if start_page < 0 or start_page >= len(reader.pages):
            raise IndexError("Start page is out of range.")

        for i in range(start_page, min(start_page + num_pages, len(reader.pages))):
            text += reader.pages[i].extract_text() or ""

    with open(output_file, "w", encoding='utf-8') as out_file:
        out_file.write(text)


pdf_path = "Harry_Potter_(www.ztcprep.com).pdf"  # Path to PDF file

# Input for Date of Birth
birth_month = int(input("Enter your birth month (1-12): "))
birth_date = int(input("Enter your birth date (1-31): "))
birth_year = int(input("Enter your birth year (e.g., 2000): "))

# Determining which book to select based on birth month
if birth_month >= 8:
    selected_book_number = math.ceil(birth_month / 2)  # Rounding up the month divided by 2
else:
    selected_book_number = birth_month  # Using the month directly for books 1-7

# Calculating page numbers based on birth date and year
start_page_file1 = birth_date - 1  # Page number corresponding to birth date (0-indexed)
start_page_file2 = 100 + (birth_year % 100)  # 100 for years 2000 onwards

# Extracting pages
extract_pages_from_pdf(pdf_path, start_page_file1, 10, "file1.txt")
extract_pages_from_pdf(pdf_path, start_page_file2, 10, "file2.txt")

print(f"Extracted pages for Book {selected_book_number} to 'file1.txt' starting from page {start_page_file1 + 1}.")
print(f"Extracted pages for year {birth_year} to 'file2.txt' starting from page {start_page_file2 + 1}.")
```

```python
import re
from collections import defaultdict

def map_reduce_word_count(file_path):
    word_count = defaultdict(int)

    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
        words = re.findall(r'\w+', text.lower())

        # Map Phase: Count occurrences
        for word in words:
            word_count[word] += 1

    return word_count

# Analyzing file1.txt
file1_word_count = map_reduce_word_count("file1.txt")

# Print word counts
for word, count in file1_word_count.items():  # Changed word1_word_count to file1_word_count
    print(f"'{word}': {count} times")
```

```python
import re
from collections import defaultdict
from spellchecker import SpellChecker

# Function to load English words
def load_english_words():
    # Initialize the SpellChecker
    spell = SpellChecker()
    return spell.word_frequency

# Function to count non-English words in a text file
def count_non_english_words(file_path, english_words):
    non_english_count = defaultdict(int)

    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read().lower()  # Read the entire file and convert to lower case
        words = re.findall(r'\w+', text)  # Find all words using regex

        for word in words:
            if word not in english_words:  # Check if the word is not in the English words
                non_english_count[word] += 1  # Increment count for non-English words

    return non_english_count


```

```python
# Load English words
english_words = load_english_words()

# Analyze file2.txt for non-English words
file2_non_english_count = count_non_english_words("file2.txt", english_words)

# Display the non-English word counts
print("Non-English word occurrences in file2.txt:")
for word, count in file2_non_english_count.items():
    print(f"'{word}': {count} times")
```
